```python
import sqlite3
```

```python
airports = pd.read_csv('bootcamp-katowice1/Pandas/nycflights13_airports.csv.gz',comment='#')
flights = pd.read_csv('bootcamp-katowice1/Pandas/nycflights13_flights.csv.gz',comment='#')
planes = pd.read_csv('bootcamp-katowice1/Pandas/nycflights13_planes.csv.gz',comment='#')
weather = pd.read_csv('bootcamp-katowice1/Pandas/nycflights13_weather.csv.gz',comment='#')
```

```python
con = sqlite3.connect('nycflights13.db')

planes.to_sql("planes", con, if_exists="replace")
airlines.to_sql("airlines", con, if_exists="replace")
flights.to_sql("flights", con, if_exists="replace")
airports.to_sql("airports", con, if_exists="replace")
weather.to_sql("weather", con, if_exists="replace")
```

```python
sql_result = pd.read_sql_query("SELECT DISTINCT engine FROM planes",con)
sql_result
```

```python
pd.DataFrame(planes.engine.unique())
```

```python
pd.read_sql_query("SELECT DISTINCT type, manufacturer FROM planes",con)
```

```python
planes[['type', 'manufacturer']].drop_duplicates().reset_index(drop=True)
```

```python
pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine",con)
```

```python
pd.DataFrame(planes.engine.value_counts().sort_index()).reset_index().rename({"index":"engine", "engine":"COUNT(*)"},axis=1)[['COUNT(*)','engine']]
```

```python
pd.pivot_table(planes[['engine']],columns=['engine'], aggfunc='count')
```

```python
pd.DataFrame(planes.groupby('engine').size())
```

```python
pd.read_sql_query('SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type',con)
```

```python
pd.DataFrame(planes.groupby(['engine','type']).size())\
                    .reset_index().rename({0:'COUNT(*)'},axis=1)[['COUNT(*)','engine','type']]
```
