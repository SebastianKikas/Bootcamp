```python
conn = sqlite3.connect('chinook.db')
c = conn.cursor()
```

```python
c.execute('SELECT title FROM albums ORDER BY title')
pprint(c.fetchall())
```

```python
c.execute('SELECT LastName, FirstName, title FROM employees ORDER BY LastName DESC')
pprint(c.fetchall())
```

```python
#Ile albumów jest dostępnych w sklepie?
c.execute('SELECT count(*) FROM albums')
pprint(c.fetchall())
```

```python
#Jaka jest średnia długość utworu sprzedawanego w sklepie?
c.execute('SELECT avg(milliseconds) FROM tracks')
pprint(c.fetchall())
```

```python
#Ile trwa najkrótszy utwór w dostępny w sklepie sklepie?
c.execute('SELECT min(milliseconds) FROM tracks')
pprint(c.fetchall())
```

```python
#Jaki jest najdłuższy utwór w dostępny w sklepie?
c.execute('SELECT name, max(milliseconds) FROM tracks')
pprint(c.fetchall())
```

```python
#Utwórów którego artysty jest najwięcej w sklepie
c.execute('Select max(cou) From (SELECT composer, count(composer) as cou FROM tracks  GROUP BY composer ORDER BY count(composer) DESC)')
pprint(c.fetchall())
```

```python
#Który pracownik opiekuje się największą liczbą klientów
c.execute('SELECT EmployeeId, max(cou) FROM (SELECT EmployeeId, count(EmployeeId) as cou FROM employees JOIN customers ON employees.EmployeeId=customers.supportrepID GROUP BY EmployeeId)')
pprint(c.fetchall())
```

```python
#Ile wynosi najwyższa kwota na fakturze
c.execute('SELECT max(Total) FROM invoices')
pprint(c.fetchall())
```
